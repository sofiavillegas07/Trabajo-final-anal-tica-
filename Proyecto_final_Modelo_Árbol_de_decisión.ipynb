0. Cargar las librerias de referencia para el modelo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #graficador
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

from google.colab import drive
drive.mount('/content/drive')
```

1. Se procede con la carga de los datos del archivo

```python
nxl='/content/drive/MyDrive/Analitica de negocios /Base de Datos_Final.xlsx'
XDB=pd.read_excel(nxl,sheet_name=0)
XDB=XDB.dropna() #Elimina celdas vacias sin datos
XDB.head(100)

categorical_columns = ['Operating System', 'Gender', 'Device Model']
label_encoder = LabelEncoder()

for column in categorical_columns:
    XDB[column] = label_encoder.fit_transform(XDB[column])

#Se procede con la seleccion de las variabales de trabajo
XD=np.array(XDB[['App Usage Time (min/day)','Screen On Time (hours/day)','Battery Drain (mAh/day)','Number of Apps Installed','Data Usage (MB/day)','Age',]]) #Datos de entrada
print(XD)
yd=np.array(XDB[['User Behavior Class']]) #Este es el dato que quiero que el modelo pronostique
print(yd)
```

2. Creación del modelo - Decision Tree

```python
mar=DecisionTreeClassifier(criterion='gini',max_depth=5) #4 niveles o variables de entrada.
mar.fit(XD,yd)
ydp=mar.predict(XD)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(yd,ydp)
print(cm)

#Evaluamos el modelo
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1]

Ex=(VN+VP)/(VN+VP+FN+FP) #Que porcentaje de los datos pronostico correctamente
print("La exactitud alcanzada por el modelo es:", Ex)
Se=VP/(FN+VP) #Que porcentaje de los datos positivos pronosticado correctamente
print("La sensibilidad alcanzada por el modelo es:", Se)
Es=VN/(VN+FP) #Que porcentaje de los datos positivos pronosticado correctamente
print("La especificidad alcanzada por el modelo es:", Es)
```

3. Graficar el arbol

```python
#Estas son las librerias para graficar el arbol
from sklearn.tree import export_graphviz
from pydotplus import graph_from_dot_data

vs=['App Usage Time (min/day)','Screen On Time (hours/day)','Battery Drain (mAh/day)','Number of Apps Installed','Data Usage (MB/day)','Age'] #son los titulos para los nodos
dot_graph=export_graphviz(mar,feature_names=vs)
graph=graph_from_dot_data(dot_graph)
graph.write_png('arbol.png')
```

4. Se procede con el pronostico

```python
XDB2=pd.read_excel(nxl,sheet_name=0)
XDB2=XDB2.dropna() #Elimina celdas vacias sin datos
XDB2.head(100)

XD2=np.array(XDB2[['App Usage Time (min/day)','Screen On Time (hours/day)','Battery Drain (mAh/day)','Number of Apps Installed','Data Usage (MB/day)','Age']]) #Datos de entrada
print(XD2)

#Pronostico: User Behavior Class
ydp2=mar.predict(XD2)
print(ydp2)
print("User behavior class:",np.sum(ydp2))
print("El numero total de datos es:",len(ydp2))

```
